```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = 'C:/Users/ahmed/Downloads/movies.csv'
data = pd.read_csv(path)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['year'] = data['year'].astype(int)
```

```python
data['year'].dtype
```

```python
# Finding Correlation
numerical_attributes = ['votes', 'budget', 'gross', 'score']
data[numerical_attributes].hist(figsize = (15, 6), color = 'blue', edgecolor = 'red', layout = (2, 2));
```

```python
top_voted = data.sort_values(['votes'], ascending = False)
```

```python
# Top voted movies
fig,axs=plt.subplots(figsize=(20,10))
g=sns.barplot(x=top_voted['name'][:7],y=top_voted['votes'][:7], palette = 'hls')
g.set_title("Top Voted Movies", weight = "bold")
plt.show()
```

```python
# Top voted movies ratings
top_imdb = data.sort_values(['score'], ascending = False)
fig,axs=plt.subplots(figsize=(20,10))
g=sns.barplot(x=top_imdb['name'][:7],y=top_imdb['score'][:7], palette = 'husl')
g.set_title("IMDB Rating of top voted movies", weight = "bold")
plt.show()
```

```python
# Gross of top rated movies
fig,axs=plt.subplots(figsize=(20,10))
g=sns.barplot(x=top_imdb['name'][:7],y=top_imdb['gross'][:7], palette = 'husl')
g.set_title("Gross by top rated movies", weight = "bold")
plt.show()
```

```python
# Top grossing movies
highest_earning = data.sort_values(['gross'], ascending = False)
fig,axs=plt.subplots(figsize=(25,10))
g=sns.barplot(x=highest_earning['name'][:7],y=highest_earning['gross'][:7], palette = 'husl')
g.set_title("Movies with highest Gross", weight = "bold")
plt.show()
```

```python
# Most Appereance by a director
fig,axs=plt.subplots(figsize=(20,5))
g=sns.barplot(x=data['director'].value_counts()[:10].index,y=data['director'].value_counts()[:10])
g.set_title("Mostly Occurred Directors", weight = "bold")
g.set_xlabel("Directors", weight = "bold")
plt.show()
```

```python
# Most Appearances by an Actor
fig,axs=plt.subplots(figsize=(20,7))
g=sns.barplot(x=data['star'].value_counts()[:10].index,y=data['star'].value_counts()[:10])
g.set_title('Most Appearances by an Actor')
g.set_ylabel("Appearances", weight = "bold")
plt.show()
```
